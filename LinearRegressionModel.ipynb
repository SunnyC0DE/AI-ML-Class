```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
```

```python
class LinReg:
  def __init__(self):
    self.c = None
    self.m = None

  def fitt(self,x , y):
    self.x = np.array(x)
    self.y = np.array(y)
    num = sum((self.x - self.x.mean())*(self.y - self.y.mean()))
    den = sum((self.x - self.x.mean())**2)
    m = num / den
    c = self.y.mean() - m* self.x.mean()
    self.m = m
    self.c = c
    self.y = self.m * self.x.mean() + self.c


  def predd(self,x):
    x = np.array(x)
    self.y = self.m * x.mean() + self.c
    return self.y
```

```python
le = LinReg()
```

```python
x,y = make_regression(n_samples=200, n_features=1, n_informative=1, n_targets=1)
```
